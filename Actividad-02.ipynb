```python
# CÃ¡lculo de autovalores y autovectores

import numpy as np

# Definimos la matriz
A = np.array([[5, 1],
              [2, 2]])

# Calculamos autovalores y autovectores
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Autovalores:", eigenvalues)
print("Autovectores:\n", eigenvectors)
```

```python
# Realiza su autodescomposiciÃ³n y grafica los resultados obtenidos en un plano con Matplotlib.

import matplotlib.pyplot as plt

# Matriz A
A = np.array([[5, 1], [2, 2]])

# Autovalores y autovectores
eigenvalues, eigenvectors = np.linalg.eig(A)

# Origen para cada flecha
origin = np.zeros((2, eigenvectors.shape[1]))

# Crear figura
plt.figure(figsize=(6, 6))

# Graficar autovectores
plt.quiver(*origin, eigenvectors[0], eigenvectors[1],
           angles='xy', scale_units='xy', scale=1,
           color=['r', 'b'], label='Autovectores')

# Graficar transformaciÃ³n de la base estÃ¡ndar
x = np.array([[1, 0], [0, 1]])
transformed = A @ x.T
plt.quiver(*origin, transformed[0], transformed[1],
           angles='xy', scale_units='xy', scale=1,
           color=['g', 'purple'], alpha=0.5, label='Base transformada')

# EstÃ©tica del grÃ¡fico
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("Autovectores y transformaciÃ³n de la base")
plt.gca().set_aspect('equal')
plt.legend()

# Guardar imagen
plt.savefig("autovectores_matriz_A.png", dpi=300)
plt.show()
```

```python
import numpy as np

v1 = np.array([3, 4])
v2 = np.array([-2, 5])

# Producto punto
dot_product = np.dot(v1, v2)

# Normas
norm_v1 = np.linalg.norm(v1)
norm_v2 = np.linalg.norm(v2)

# Ãngulo en radianes
theta_rad = np.arccos(dot_product / (norm_v1 * norm_v2))

# Ãngulo en grados
theta_deg = np.degrees(theta_rad)

print("Ãngulo en radianes:", theta_rad)
print("Ãngulo en grados:", theta_deg)
```

```python
# ProyecciÃ³n ortogonal de un vector sobre un subespacio
# Supongamos que: u = [-2, 7] y x = [3, 1]


import numpy as np

u = np.array([-2, 7])
x = np.array([3, 1])

# ProyecciÃ³n ortogonal de x sobre el subespacio generado por u
projection = (np.dot(x, u) / np.dot(u, u)) * u
print("Ï_U(x) =", projection)

##############################################################################
# VisualizaciÃ³n de la proyecciÃ³n ortogonal
import numpy as np
import matplotlib.pyplot as plt

# Vector base del subespacio
u = np.array([-2, 7])
# Vector a proyectar
x = np.array([3, 1])

# CÃ¡lculo de la proyecciÃ³n ortogonal
lambda_proj = np.dot(x, u) / np.dot(u, u)
projection = lambda_proj * u

# Graficamos
origin = np.array([[0, 0]])
fig, ax = plt.subplots()

# Vector original x
ax.quiver(*origin.T, x[0], x[1], angles='xy', scale_units='xy', scale=1, color='blue', label='x')

# Vector base u
ax.quiver(*origin.T, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='green', label='u')

# ProyecciÃ³n Ï_U(x)
ax.quiver(*origin.T, projection[0], projection[1], angles='xy', scale_units='xy', scale=1, color='red', label='Ï_U(x)')

# LÃ­nea desde x hasta su proyecciÃ³n
ax.plot([x[0], projection[0]], [x[1], projection[1]], 'k--', linewidth=1)

# ConfiguraciÃ³n del grÃ¡fico
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 8)
ax.set_aspect('equal')
ax.grid(True)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.set_title("ProyecciÃ³n ortogonal de x sobre el subespacio generado por u")
ax.legend()
plt.show()
```

```python
# matriz de proyecciÃ³n PÏ para proyectar por el subespacio

import numpy as np

u = np.array([-2, 7]).reshape(-1, 1)  # vector columna
x = np.array([3, 1]).reshape(-1, 1)

# Matriz de proyecciÃ³n
P_pi = (u @ u.T) / (u.T @ u)

# ahora encontramos la proyecciÃ³n del vector x sobre el vector u multiplicando la matriz de proyecciÃ³n por el vector x
projection = P_pi @ x

print("Matriz de proyecciÃ³n PÏ:\n", P_pi)
print("Ï_U(x):\n", projection.flatten())
```

```python
# Genera una matriz cuadrada con ( n > 3 ) y calcula su determinante y su traza.

import numpy as np

# TamaÃ±o de la matriz (n > 3)
n = 4

# Generar matriz cuadrada aleatoria
A = np.random.randint(-10, 10, size=(n, n))
print("Matriz A:\n", A)

# Calcular determinante
det_A = np.linalg.det(A)
print("\nDeterminante de A:", det_A)

# Si el determinante es distinto de cero, la matriz es invertible, lo que implica que la transformaciÃ³n es reversible.
# El determinante permite saber si un sistema de ecuaciones tiene soluciÃ³n Ãºnica (si es distinto de cero).

# Calcular traza (suma de la diagonal)
trace_A = np.trace(A)
print("Traza de A:", trace_A)
```

```python
rango = np.linalg.matrix_rank(A)
print("Rango de A:", rango)
```
